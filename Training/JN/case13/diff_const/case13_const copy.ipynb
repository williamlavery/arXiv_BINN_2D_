```python
import numpy as np

def make_C_from_D(D, beta, n_points=1001):
    """
    Create a new function C(u) = k * D(u) such that the
    mean absolute percentage error is beta (%), and print
    MSE and absolute percentage error for u in [0, 1].

    Parameters
    ----------
    D : callable
        Original function D(u).
    beta : float
        Target absolute percentage error in percent.
    n_points : int
        Number of points in [0, 1] to use for error calculation.

    Returns
    -------
    C : callable
        New scaled function C(u) = k * D(u).
    """
    # Scale factor so that APE = beta (%)
    k = 1.0 + beta / 100.0  # could also use 1 - beta/100.0

    def C(u):
        return k * D(u)

    # Sample u in [0, 1]
    u = np.linspace(0.0, 1.0, n_points)
    D_vals = np.array([D(ui) for ui in u])
    C_vals = k * D_vals

    # Mean Squared Error over [0, 1]
    mse = np.mean((C_vals - D_vals) ** 2)

    # Mean Absolute Percentage Error (%)
    # avoid division by zero if D(u) = 0 somewhere
    mask = D_vals != 0
    if np.any(mask):
        ape = np.mean(
            np.abs((C_vals[mask] - D_vals[mask]) / D_vals[mask])
        ) * 100.0
    else:
        ape = np.nan  # undefined if D(u) is identically zero

    print(f"Scale factor k = {k}")
    print(f"MSE over [0, 1]: {mse}")
    print(f"Mean absolute percentage error over [0, 1]: {ape}")

    return C, mse, ape

# From file path info contstruct true D

# From file path info contstruct true D
# --- Diffusion models ---
diffusion_models = {
    "const":     (lambda u, a: np.full_like(u, a), [0.02472]),
    "linear":    (lambda u, a, b: a + b * u,         [0.015, 0.06]),
    "quadratic": (lambda u, a, b: a + b * u**2,      [0.01, 0.044]),
    "exp":       (lambda u, a, b, c: a + b * (1 - sp.exp(-c * u)), [0.003, 0.095, 2.5])
}

# --- Growth models ---
growth_models = {
    "const":     (lambda u, a: np.full_like(u, a) / 2,          [1.3]),
    "linear":    (lambda u, a, b: (a + b * u) / 2,              [2.4, -3]),
    "quadratic": (lambda u, a, b: (a + b * u**2) / 2,           [2.1, -0.29]),
    "exp":       (lambda u, a, b, c: (a + b * (1 - sp.exp(-c * u))) / 2, [0.7, 1.3, -4]),
    "zero":      (lambda u, _: np.full_like(u, 0.0),            [0])
}

# --- Function dispatch ---
diff_func, theta_D = diffusion_models[diff_label]
grow_func, theta_G = growth_models[grow_label]

def diffusion_func(u):
    return diff_func(u, *theta_D)

def growth_func(u):
    return grow_func(u, *theta_G)


    #print("check PE:", np.mean(100*np.abs(alpha - D_true)/D_true))
mses = []
beta_list = [1,2]
for beta in beta_list:
    _,mse, ape =  make_C_from_D(diffusion_func, beta=beta)
    mses.append(mse)



name_full_D = os.path.join(name_intro_D, f'NNu64_{grow_label}_dataGamma{gamma}_D_loss_bars.png')
name_full_G = os.path.join(name_intro_G, f'NNu64_{grow_label}_dataGamma{gamma}_G_loss_bars.png')

D_colors_upd = [D_colors[3], D_colors[2],D_colors[1],D_colors[0]]
G_colors_upd = [G_colors[3], G_colors[2],G_colors[1],G_colors[0]]

plot_params_D = {
    f"{key}": [D_colors[i], linestyles[i], marker_styles[i], binn_ES_list_combo[i]]
    for i, key in enumerate(binn_ES_list_combo)
}

plot_params_G = {
    f"{key}": [G_colors[i], linestyles[i], marker_styles[i],binn_ES_list_combo[i]]
    for i, key in enumerate(binn_ES_list_combo)
}



ES_reduced = [500, 1000,2000,3000]


binn_models_combined_reduced = {
    str(ES_reduced[i]): width_dict
    for i in range(1, len(ES_reduced))
    for _, width_dict in binn_models_by_ES[i].items()
}

binn_ES_list_reduced_combo = list(binn_models_combined_reduced.keys())


plot_diff_mse_lst(models_dics_list=[binn_models_combined_reduced],
                  plot_params=plot_params_D,
                  label_list=[f"{ES}" for ES in binn_ES_list_reduced_combo],
                  name=name_full_D,
                  figsize=(7, 5),
                  bbox_to_anchor=(0.182, 1.01),
                  legend_size=10,
                  legend_bool = False,
                  colors = D_colors_upd,
                  pe_bbox_to_anchor_offset=(-0.5, 0.25),
                  pe_label_list=[f"{beta}" for beta in beta_list],
                  pe_values=None,#mses,
                  pe_linestyles=(":", "--","-", "-."),
                  pe_legend_ncols=4,
                  x_label=r"ES")


# mses = []
# beta_list = [5]
# for beta in beta_list:
#     _,mse, ape =  make_C_from_D(growth_func, beta=beta)
#     mses.append(mse)


plot_grow_mse_lst(models_dics_list=[binn_models_combined_reduced],
                  plot_params=plot_params_G,
                  label_list=[f"{ES}" for ES in binn_ES_list_reduced_combo],
                  name=name_full_G,
                  figsize=(7, 5),
                  bbox_to_anchor=(0.175, 1),
                  legend_size=10,
                  legend_bool = False,
                  colors = G_colors_upd,

                  pe_bbox_to_anchor_offset=(-0.5, 0.25),
                  pe_label_list=[f"{beta}" for beta in beta_list],
                  pe_values=None,#mses,
                  pe_linestyles=(":", "--","-", "-."),
                  pe_legend_ncols=4,
                  x_label=r"ES")


plot_dual_mse_lst(
    grow_models_dics_list=[binn_models_combined_reduced],   # growth input
    diff_models_dics_list=[binn_models_combined_reduced],   # diffusion input

    plot_params={"grow": plot_params_G, "diff": plot_params_D},

    label_list=[f"{ES}" for ES in binn_ES_list_reduced_combo],

    # -------- Appearance --------
    figsize=(7, 5),
    bbox_to_anchor=(0.18, 1.01),  # matching diffusion call
    legend_size=10,
    legend_bool=False,

    # Colors
    growth_colors=G_colors_upd,
    diffusion_colors=D_colors_upd,

    # Hatching

    # X label
    x_label=r"ES",
    x_label_fontsize=14,
    y_label_right=r"Growth MSE [day$^{-2}$]",
    
    hatch_patterns=(
                                    "xxx", "", "////", "\\\\\\\\", "xxxx", "++++", "....", "ooo",
                                 ),


    # -------- Output --------
    name=os.path.join(f"../../pngs/2D/Case13/plot_varyD/diff_{diff_label}/",
                     f'NNu64_{grow_label}_dataGamma{gamma}_DG_loss_bars.png'),

    group_width=0.8,          # total width for all ES groups at one x-label
    intra_metric_gap_factor=0.1,  # relative gap between growth & diffusion in a group

    diffusion_ref = [5.964e-7, D_colors_upd[-1]],
    growth_ref = [0.00059568, G_colors_upd[-1]]
)
```

```python


def plot_repeats_across_ES(
    dn_models_dics_list,
    ES_list,
    use_val_loss=True,
    yscale="log",
    xscale=None,
    ylim_total=(1e-2, 1e2),
    ylim_data=(1e-2, 1e2),
    ylim_pde=(1e-2, 1e2),
    figsize=(9, 5),
    base_cmap="Blues",
    es_marker_map=None,
    title=None,
    xlim_left=(1, 300),
    xlim_right=(800, 2000),
    base_dir = None,
):
    """
    Makes THREE broken-axis figures:
      1) Total validation loss (val_loss_list or train_loss_list)
      2) Validation data loss (val_data_loss_list)
      3) Validation PDE loss (val_pde_loss_list)

    dn_models_dics_list: list of dn_models_dics, one per ES
        [
          {'green_...': {0: ModelWrapper, 1: ..., 4: ...}},   # ES_list[0]
          {'green_...': {0: ModelWrapper, 1: ..., 4: ...}},   # ES_list[1]
          ...
        ]
    ES_list: list of ES values, e.g. [500, 1000, 2000]
    """

    assert len(dn_models_dics_list) == len(ES_list), "ES_list must match dn_models_dics_list length"

    # ---------- ES → marker mapping ----------
    if es_marker_map is None:
        default_markers = ["o", "s", "^", "D", "v"]
        es_marker_map = {
            ES: default_markers[i % len(default_markers)]
            for i, ES in enumerate(ES_list)
        }

    # ---------- ES → marker mapping ----------
    if es_marker_map is None:
        default_markers = ["o", "s", "^", "D", "v"]
        es_marker_map = {
            ES: default_markers[i % len(default_markers)]
            for i, ES in enumerate(ES_list)
        }

    # ---------- NEW: Marker sizes based on SHAPE ----------
    # You can modify these as needed
    marker_size_by_shape = {
        "o": 100,   # circle
        "s": 50,    # square
        "^": 40,    # triangle up
        "v": 40,    # triangle down
        "D": 60,    # diamond
    }

    # Construct ES → size mapping from ES → marker mapping
    es_marker_size = {
        ES: marker_size_by_shape.get(es_marker_map[ES], 60)   # default size if shape missing
        for ES in ES_list
    }


    # ---------- Single label assumption ----------
    first_dic = dn_models_dics_list[0]
    if len(first_dic) != 1:
        raise ValueError("Expected each dn_models_dics to contain one label.")
    base_label = list(first_dic.keys())[0]

    # ---------- Repeats ----------
    repeat_indices = sorted(first_dic[base_label].keys())
    n_repeats = len(repeat_indices)

    # consistent colours per repeat
    cmap = cm.get_cmap(base_cmap, n_repeats + 2)
    #repeat_colors = {r: cmap(r + 1) for r in repeat_indices}

    repeat_colors = {
    r: tableau_colors[r % len(tableau_colors)]
    for r in repeat_indices
    }



    # ----------------------------------------------------------------------
    # Helper to create a broken x-axis figure and return (fig, axL, axR)
    # ----------------------------------------------------------------------
    def _make_broken_axes(figsize_local):
        fig, (axL, axR) = plt.subplots(
            1, 2, figsize=figsize_local, sharey=True,
            gridspec_kw={"wspace": 0.05}
        )

        # hide right spine on left axis and left spine on right axis
        axL.spines["right"].set_visible(False)
        axR.spines["left"].set_visible(False)
        axR.yaxis.set_tick_params(labelleft=False)

        # diagonal cut marks
        d = .012
        kwargs_L = dict(transform=axL.transAxes, color='k', clip_on=False)
        axL.plot((1-d, 1+d), (0-d, 0+d), **kwargs_L)
        axL.plot((1-d, 1+d), (1-d, 1+d), **kwargs_L)

        kwargs_R = dict(transform=axR.transAxes, color='k', clip_on=False)
        axR.plot((-d, +d), (1-d, 1+d), **kwargs_R)
        axR.plot((-d, +d), (0-d, 0+d), **kwargs_R)

        return fig, axL, axR

    # ----------------------------------------------------------------------
    # Shared legend handles (same for all three figures)
    # ----------------------------------------------------------------------
    repeat_handles = [
        Line2D([0], [0], color=repeat_colors[r], lw=2, label=f"repeat {r}")
        for r in repeat_indices
    ]

    es_handles = [
        Line2D(
            [0], [0],
            marker=es_marker_map[ES],
            linestyle="",
            markeredgecolor="k",
            markerfacecolor="gray",
            markersize=np.sqrt(es_marker_size[ES]),  # legend approx
            label=f"ES = {ES}",
        )
        for ES in ES_list
    ]

    # ----------------------------------------------------------------------
    # 1) TOTAL LOSS FIGURE
    # ----------------------------------------------------------------------
    fig1, axL1, axR1 = _make_broken_axes(figsize)

    for dn_models_dics, ES in zip(dn_models_dics_list, ES_list):
        model_dic = dn_models_dics[base_label]
        marker = es_marker_map[ES]
        msize = es_marker_size[ES]

        for r in repeat_indices:
            wrapper = model_dic[r]

            # total (val or train) loss
            if use_val_loss and len(wrapper.val_loss_list):
                losses = np.array(wrapper.val_loss_list, float)
            else:
                losses = np.array(wrapper.train_loss_list, float)

            epochs = np.arange(len(losses))
            color = repeat_colors[r]

            # plot lines (low alpha to make markers stand out)
            for ax in (axL1, axR1):
                ax.plot(
                    epochs, losses,
                    lw=1, alpha=0.5,
                    color=color
                )

            # early stopping index based on total validation loss
            best_loss = wrapper.best_val_loss
            idx_es = int(np.argmin(np.abs(losses - best_loss)))

            for ax in (axL1, axR1):
                ax.scatter(
                    epochs[idx_es], losses[idx_es],
                    marker=marker,
                    s=msize,
                    edgecolor="k",
                    linewidth=0.75,
                    color=color,
                    zorder=4
                )

    # style
    axL1.set_xlim(*xlim_left)
    axR1.set_xlim(*xlim_right)
    if yscale:
        axL1.set_yscale(yscale)
    if xscale:
        axL1.set_xscale(xscale)
        axR1.set_xscale(xscale)
    if ylim_total:
        axL1.set_ylim(ylim_total)

    axL1.set_xlabel("Epoch")
    axR1.set_xlabel("Epoch")
    axL1.set_ylabel("Total loss")
    axL1.grid(True, ls="--", alpha=0.3)
    axR1.grid(True, ls="--", alpha=0.3)

    ttl_total = title if title is not None else f"Total loss trajectories for {base_label}"
    print(ttl_total)

    L1a = axL1.legend(
        handles=repeat_handles,
        title="Repeats (colour)",
        loc="upper right",
        fontsize=9,
        title_fontsize=10
    )
    axL1.add_artist(L1a)

    axL1.legend(
        handles=es_handles,
        title="Early stopping (marker)",
        loc="lower left",
        fontsize=9,
        title_fontsize=10
    )

    fig1.tight_layout()

    if base_dir:
        plt.savefig(os.path.join(base_dir, "total_loss.png"), dpi=100, bbox_inches='tight', facecolor='None')
    plt.show()

    # ----------------------------------------------------------------------
    # 2) DATA LOSS FIGURE
    # ----------------------------------------------------------------------
    fig2, axL2, axR2 = _make_broken_axes(figsize)

    for dn_models_dics, ES in zip(dn_models_dics_list, ES_list):
        model_dic = dn_models_dics[base_label]
        marker = es_marker_map[ES]
        msize = es_marker_size[ES]

        for r in repeat_indices:
            wrapper = model_dic[r]

            if not hasattr(wrapper, "val_data_loss_list") or len(wrapper.val_data_loss_list) == 0:
                continue  # skip if not present

            data_losses = np.array(wrapper.val_data_loss_list, float)

            # epochs length should match; but recompute to be safe
            epochs = np.arange(len(data_losses))
            color = repeat_colors[r]

            for ax in (axL2, axR2):
                ax.plot(
                    epochs, data_losses,
                    lw=1, alpha=0.5,
                    color=color
                )

            # ES index reused from total loss definition
            if use_val_loss and len(wrapper.val_loss_list):
                total_losses = np.array(wrapper.val_loss_list, float)
            else:
                total_losses = np.array(wrapper.train_loss_list, float)
            best_loss = wrapper.best_val_loss
            idx_es = int(np.argmin(np.abs(total_losses - best_loss)))

            # guard for different lengths
            idx_es = min(idx_es, len(data_losses) - 1)

            for ax in (axL2, axR2):
                ax.scatter(
                    epochs[idx_es], data_losses[idx_es],
                    marker=marker,
                    s=msize,
                    edgecolor="k",
                    linewidth=0.75,
                    color=color,
                    zorder=4
                )

    axL2.set_xlim(*xlim_left)
    axR2.set_xlim(*xlim_right)
    if yscale:
        axL2.set_yscale(yscale)
    if xscale:
        axL2.set_xscale(xscale)
        axR2.set_xscale(xscale)
    if ylim_data:
        axL2.set_ylim(ylim_data)

    axL2.set_xlabel("Epoch")
    axR2.set_xlabel("Epoch")
    axL2.set_ylabel("Data loss")
    axL2.grid(True, ls="--", alpha=0.3)
    axR2.grid(True, ls="--", alpha=0.3)

    ttl_data = f"Data loss trajectories for {base_label}"
    print(ttl_data)

    L2a = axL2.legend(
        handles=repeat_handles,
        title="Repeats (colour)",
        loc="upper right",
        fontsize=9,
        title_fontsize=10
    )
    axL2.add_artist(L2a)

    axL2.legend(
        handles=es_handles,
        title="Early stopping (marker)",
        loc="lower left",
        fontsize=9,
        title_fontsize=10
    )

    fig2.tight_layout()
    if base_dir:
        plt.savefig(os.path.join(base_dir, "data_loss.png"), dpi=100, bbox_inches='tight', facecolor='None')
    plt.show()

    # ----------------------------------------------------------------------
    # 3) PDE LOSS FIGURE
    # ----------------------------------------------------------------------
    fig3, axL3, axR3 = _make_broken_axes(figsize)

    for dn_models_dics, ES in zip(dn_models_dics_list, ES_list):
        model_dic = dn_models_dics[base_label]
        marker = es_marker_map[ES]
        msize = es_marker_size[ES]

        for r in repeat_indices:
            wrapper = model_dic[r]

            if not hasattr(wrapper, "val_pde_loss_list") or len(wrapper.val_pde_loss_list) == 0:
                continue  # skip if not present

            pde_losses = np.array(wrapper.val_pde_loss_list, float)

            epochs = np.arange(len(pde_losses))
            color = repeat_colors[r]

            for ax in (axL3, axR3):
                ax.plot(
                    epochs, pde_losses,
                    lw=1, alpha=0.5,
                    color=color
                )

            # ES index from total loss
            if use_val_loss and len(wrapper.val_loss_list):
                total_losses = np.array(wrapper.val_loss_list, float)
            else:
                total_losses = np.array(wrapper.train_loss_list, float)
            best_loss = wrapper.best_val_loss
            idx_es = int(np.argmin(np.abs(total_losses - best_loss)))
            idx_es = min(idx_es, len(pde_losses) - 1)

            for ax in (axL3, axR3):
                ax.scatter(
                    epochs[idx_es], pde_losses[idx_es],
                    marker=marker,
                    s=msize,
                    edgecolor="k",
                    linewidth=0.75,
                    color=color,
                    zorder=4
                )

    axL3.set_xlim(*xlim_left)
    axR3.set_xlim(*xlim_right)
    if yscale:
        axL3.set_yscale(yscale)
    if xscale:
        axL3.set_xscale(xscale)
        axR3.set_xscale(xscale)
    if ylim_pde:
        axL3.set_ylim(ylim_pde)

    axL3.set_xlabel("Epoch")
    axR3.set_xlabel("Epoch")
    axL3.set_ylabel("PDE loss")
    axL3.grid(True, ls="--", alpha=0.3)
    axR3.grid(True, ls="--", alpha=0.3)

    ttl_pde = f"PDE loss trajectories for {base_label}"
    print(ttl_pde)

    L3a = axL3.legend(
        handles=repeat_handles,
        title="Repeats (colour)",
        loc="upper right",
        fontsize=9,
        title_fontsize=10
    )
    axL3.add_artist(L3a)

    axL3.legend(
        handles=es_handles,
        title="Early stopping (marker)",
        loc="lower left",
        fontsize=9,
        title_fontsize=10
    )

    fig3.tight_layout()
    if base_dir:
        plt.savefig(os.path.join(base_dir, "pde_loss.png"), dpi=100, bbox_inches='tight', facecolor='None')
    plt.show()

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.cm as cm
import os

tableau_colors = [
    "#1F77B4",  # strong blue
    "#FF7F0E",  # vivid orange
    "#2CA02C",  # green
    "#D62728",  # red
    "#9467BD",  # purple
    "#8C564B",  # brown
    "#E377C2",  # pink
    "#7F7F7F",  # gray
    "#BCBD22",  # yellow/green
    "#17BECF",  # teal
]
plot_repeats_across_ES(
    [{'linear': binn_models_dics[es]['linear']} for es in ES_reduced[1:]],
    ES_list=ES_reduced[1:],
    use_val_loss=True,
    yscale="log",
    xscale="log",
    ylim_total=  (4*10**-5, 10**0),
    ylim_data= (4*10**-5, 10**0),
    ylim_pde= (10**-6, 10**-0),
    xlim_left=(1, 4000),
    xlim_right=(4000, 20000),
    figsize=(7, 5),
    base_cmap="Blues",
    es_marker_map=None,
    title=None,
    #base_dir = save_path1
)
```

```python
def ic1(x, y):
    """
    2D analogue of the updated ic1(x):
    ic1(x) = 1/2 - 1/2 * cos(...) * cos(...)

    Returns:
        ic1(x,y) = 1/2 
                   - 1/2 * cos_x^2(...) * cos_y^2(...)
    """

    mean_x = np.mean(x)
    mean_y = np.mean(y)

    Lx = np.max(x) - np.min(x)
    Ly = np.max(y) - np.min(y)

    X, Y = np.meshgrid(x, y, indexing='ij')

    cos_x = np.cos(2 * np.pi * (X - mean_x) / Lx)
    cos_y = np.cos(2 * np.pi * (Y - mean_y) / Ly)

    return 0.5 - 0.5 * (cos_x * cos_x) * (cos_y * cos_y)
```

```python
import numpy as np

def ic1(x, y, amplitude=1.0):
    """
    2D cosine-squared bump, like a Gaussian, centered at the geometric
    centre of the domain and decaying smoothly towards the boundaries.

    ic1(x, y) = amplitude * cos^2(π (x - xc) / Lx) * cos^2(π (y - yc) / Ly)

    where xc, yc are the geometric centres of the x, y domains.
    """

    x = np.asarray(x)
    y = np.asarray(y)

    # Geometric centre of the domain
    xc = 0.5 * (x.min() + x.max())
    yc = 0.5 * (y.min() + y.max())

    Lx = x.max() - x.min()
    Ly = y.max() - y.min()

    X, Y = np.meshgrid(x, y, indexing='ij')

    denomx = Lx * 0.5
    denomy = Ly * 0.5


    cos_x = np.cos(np.pi * (X - xc) / denomx)
    cos_y = np.cos(np.pi * (Y - yc) / denomy)

    return 1-   (cos_x**2) * (cos_y**2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_ic1(ic_func, x=None, y=None):
    """
    Quick 2D visualization of an IC function ic_func(x, y),
    plus fixed-x and fixed-y slices through the geometric centre.
    """
    if x is None:
        x = np.linspace(0.0, 1.0, 200)
    if y is None:
        y = np.linspace(0.0, 1.0, 200)

    Z = ic_func(x, y)  # shape (Nx, Ny)

    # Centre indices
    ix = len(x) // 2   # center x index
    iy = len(y) // 2   # center y index

    x_center = x[ix]
    y_center = y[iy]

    # Extract slices
    slice_y_fixed = Z[:, iy*0]   # Z(x, y_center)
    slice_x_fixed = Z[ix, :]   # Z(x_center, y)

    # ---- Plot 2D map ----
    plt.figure(figsize=(6, 5))
    im = plt.imshow(
        Z.T,
        origin='lower',
        extent=(x.min(), x.max(), y.min(), y.max()),
        aspect='equal'
    )
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Initial condition ic1 (2D field)")
    plt.colorbar(im, label="value")
    plt.scatter([x_center], [y_center], color='red', s=30, label="slice location")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # ---- Plot slices ----
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    # slice at fixed y
    ax[0].plot(x, slice_y_fixed, lw=2)
    ax[0].axvline(x_center, color='r', ls='--')
    ax[0].set_title(f"Slice at y = {y_center:.3f}")
    ax[0].set_xlabel("x")
    ax[0].set_ylabel("ic1(x, y_center)")

    # slice at fixed x
    ax[1].plot(y, slice_x_fixed, lw=2)
    ax[1].axvline(y_center, color='r', ls='--')
    ax[1].set_title(f"Slice at x = {x_center:.3f}")
    ax[1].set_xlabel("y")
    ax[1].set_ylabel("ic1(x_center, y)")

    plt.tight_layout()
    plt.show()

plot_ic1(ic1, x=data_obj.x1, y=data_obj.x2)
```
