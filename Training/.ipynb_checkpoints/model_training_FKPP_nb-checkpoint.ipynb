## Train & download
- `train`: Bool
              True: The BINN model is trained (either extending an existing 
                    training or carrying-out the first round of training).
              False: The BINN model is not trained.
- `download`: Bool
              True: The .../Notebooks/FKPP/Models directory is downloaded to the 
                    local disk (overwritting the existing corresponding 
                    directory).
              False: .../Notebooks/FKPP/Models directory is not downloaded.

```python
# USER INPUT  
train = True
download = True
```

```python
server_obj.remote_cwd
```

```python
sm.server_commands([f'cd {server_obj.remote_cwd}/Notebooks/Training/FKPP/Models/','ls'])
```

```python
server_obj.remote_trained_models_dir_parent,os.getcwd()
```

```python
sm.download_dir_from_server(server_obj.remote_trained_models_dir_parent,
                            os.getcwd())
```

```python
for _ in range(1):


    if train: 

        model_num = server_obj.model_params['model_num']
        delay = server_obj.model_params['delay']
        IC = server_obj.model_params['IC']

        
        ex.choose_model(model_num, IC, delay, server_obj.dataobj)  # Model to be trained
        
        
        save_path_model_remote = f"{server_obj.remote_cwd}/{data_obj.save_path_model_relative}"
        
        sm.add_file_to_server(server_obj.dataobj.save_path_model_relative +'/model_info.txt',
                              save_path_model_remote)

        print('++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
        print(f'-------------------- model_num = {model_num} | IC = {IC} | delay = {delay} ------------------\n')
        print('++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')

        if train:
            print("\nThe training process begins... \n")
            print(f"Current date and time: {datetime.now()}\n")
            _, new_epochs, _, _, _ = ex.chosen_model(server_obj.dataobj)

            print(f"Estimated finish time: {datetime.now() + timedelta(seconds=new_epochs/10 * 16)}\n")

            sm.server_commands([f'cd {server_obj.remote_cwd}',
                                'source ~/miniconda3/etc/profile.d/conda.sh',
                                f'conda activate {server_obj.conda_environ}',
                                f'python model_training_FKPP_comparison.py'])

    if download:
        sm.download_dir_from_server(server_obj.remote_trained_models_dir_parent,
                            os.getcwd())
```
